```python
#QR CODE GENERATOR
import qrcode
def create_qr(data, filename="my_insta.jpeg"):
    qr = qrcode.QRCode(
        version=1,
        box_size=10,
        border=4
    )
    
    qr.add_data(data)
    qr.make(fit=True)
    
    img = qr.make_image(
        fill_color="black",
        back_color="white"
    )
    
    img.save(filename)
    print(f"QR saved as {filename}")

create_qr(
    "https://www.instagram.com/muskan_nanwa?igsh=am1rN2JmZTRwcDQ1"
)
```
